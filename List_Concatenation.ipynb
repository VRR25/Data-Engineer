```python
# create the input lists

first_name = ["Akesh", "Mukesh","Ramesh","Sukesh"] 
last_name = ["Sai","Babu","Damu","Reddy"]

# create the empty list which contains the full name
full_name=[]

print("Names before appending :" + str(first_name) + " and " + str(last_name))

# use for loop for iterating through each element of each list

for i in range(len(first_name)):
    full_name.append(first_name[i] + ' ' + last_name[i])

print("Names after appending : " + str(full_name))

    
```
